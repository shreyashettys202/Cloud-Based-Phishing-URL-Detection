```python
import pandas as cloudUrlpan
import seaborn as cloudUrlseabn
import matplotlib.pyplot as cloudUrlmpt
import nltk as cloudNlt
cloudNlt.download('stopwords')
import regex as cloudRgx
from nltk.corpus import stopwords as cloudStp
from sklearn import preprocessing as cloudprlb

#====# Read Phishing URL data
cloudUrlData = cloudUrlpan.read_csv('phishing_site_urls.csv')
cloudUrlData

#====# Displaying Output URL types
print("\nCloud Phishing URL Categories\n", cloudUrlData['Label'].value_counts())
cloudFurl = cloudUrlmpt.figure(figsize=(4, 3), facecolor='lightgrey')
cloudUrlseabn.countplot(y ='Label', data = cloudUrlData, color = 'purple')
cloudUrlmpt.title('\nCloud - Phishing URL Types')
cloudUrlmpt.show()

#====# Checking missing values and duplicates
print("\nNumber of Missing Values - Cloud:",cloudUrlData.isnull().values.sum())
print("\nNumber of Duplicates - Cloud    :",cloudUrlData[cloudUrlData.duplicated()].shape[0])
cloudUrlData = cloudUrlData.drop_duplicates()

#====# Stopwords removal
cloudStp = cloudStp.words('english')
def detect(cloudph):
    cloudph = cloudph.lower()
    cloudph = cloudRgx.sub("[^a-z]+", " ", cloudph)
    cloudph = " ".join([d for d in cloudph.split() if d not in cloudStp])
    return cloudph

cloudphishdet = []
for c in cloudUrlData['URL']:
    cloudphishdet.append(detect(c))
cloudUrlData['URL'] = cloudphishdet
cloudUrlData

#====# object to numeric changes
cloudlbl = cloudprlb.LabelEncoder()
cloudUrlData['Label']= cloudlbl.fit_transform(cloudUrlData['Label'])
print("\nCloud Phishing URL Categories\n",cloudUrlData['Label'].value_counts())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer as cloudTf
from sklearn.decomposition import TruncatedSVD as cloudSvd
from imblearn.over_sampling import SMOTE as cloudSmte
from collections import Counter as cloudCn
from sklearn.metrics import confusion_matrix as cloudCf
from sklearn.metrics import ConfusionMatrixDisplay as cloudCd
from sklearn.metrics import classification_report as cloudCr
from sklearn.model_selection import train_test_split as cloudTTsp

cloudUrlX = cloudUrlData['URL']
cloudUrlY = cloudUrlData['Label']

#====# TF-IDF
cloudModUr = cloudTf(analyzer='word', ngram_range=(1, 1))
cloudUrlX = cloudModUr.fit_transform(cloudUrlX)
print("URLs after TFIDF: ", cloudUrlX.shape)

#====# Truncated SVD
cloudModUr = cloudSvd(n_components=12)
cloudUrlX = cloudModUr.fit_transform(cloudUrlX)
print("\nURL Shape after SVD: ", cloudUrlX.shape)


#====# SMOTE Balancing
print('\nUnbalanced URLs %s' % cloudCn(cloudUrlY))
cloudModUr = cloudSmte()
cloudUrlXre, cloudUrlYre = cloudModUr.fit_resample(cloudUrlX, cloudUrlY)
print('\nBalanced URLs %s' % cloudCn(cloudUrlYre))

#====# Convert the resampled arrays
Xcloud = cloudUrlpan.DataFrame(cloudUrlXre, columns=[f'component_{i}' for i in range(cloudUrlXre.shape[1])])
Ycloud = cloudUrlpan.Series(cloudUrlYre, name='Label')

##====# Concatenating input and output
cloudUrlData = cloudUrlpan.concat([Xcloud, Ycloud], axis=1)
#====# Resampled DataFrame
print("\nShape of Balanced Data: ", cloudUrlData.shape)
print("\nURL Types:\n", cloudUrlData['Label'].value_counts())


Xtrn_Cloud, Xtst_Cloud, Ytrn_Cloud, Ytst_Cloud = cloudTTsp(Xcloud, Ycloud, test_size=0.20, random_state= 15)

cloudTr, cloudTs = Xtrn_Cloud.shape, Xtst_Cloud.shape
print(f"{cloudTr} are 80% Training, and {cloudTs} are 20% Testing")
```

```python
from sklearn.neural_network import MLPClassifier as cloudMlpClf
import joblib as cloudjobl

cloudModUr = cloudMlpClf()
cloudModUr.fit(Xtrn_Cloud, Ytrn_Cloud)
YcloudPr = cloudModUr.predict(Xtst_Cloud)
print(cloudCr(Ytst_Cloud, YcloudPr))
cloudMat = cloudCf(Ytst_Cloud, YcloudPr)
cloudCd(confusion_matrix = cloudMat, display_labels = [0, 1]).plot(cmap='Blues')
cloudUrlmpt.show()

cloudMSave = cloudjobl.dump(cloudModUr, 'neuralnetwork.sav')
```
